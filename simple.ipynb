```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x=[1,2,3,4,5,6]
y=[10,20,10,30,25,80]

df=pd.DataFrame({
    'days':x,
    'No. of People':y
})

df.head(6)
```

```python
print(df['days'].unique())
```

```python
sns.lineplot(x='days',y='No. of People', data=df)
plt.show()
```

```python
# Day 2  "Bar Plot in Seaborn"

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df2=sns.load_dataset('titanic')
df2.head()
```

```python
df2.shape
```

```python
df2['who'].value_counts()
```

```python
sns.barplot(x='who',y='fare',data=df2,color='red',legend=True, palette=["red", "green", "blue"],saturation =0.2,ci=False)
```

```python

plt.bar(df2['who'],df2['fare'])
```

```python
grouped=df2.groupby('who')['fare'].mean()

plt.bar(grouped.index,grouped.values)
plt.show()
```

```python
from numpy import median
sns.barplot(x='fare',y='who',data=df2,facecolor=[1,1,1,0],edgecolor='0.2')
```

```python
# set the order of each and every bar in the barplot
sns.barplot(x='who',y='fare',data=df2,order=['woman','child','man'],color='orange')
```
