```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x=[1,2,3,4,5,6]
y=[10,20,10,30,25,80]

df=pd.DataFrame({
    'days':x,
    'No. of People':y
})

df.head(6)
```

```python
print(df['days'].unique())
```

```python
sns.lineplot(x='days',y='No. of People', data=df)
plt.show()
```
